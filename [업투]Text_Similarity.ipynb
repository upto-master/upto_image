```python
from konlpy.tag import Okt
from tqdm import tqdm
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#okt 객체 선언
okt = Okt() 
noun_list = []
for content in tqdm(text_df['이미지 설명']): 
    nouns = okt.nouns(content) # 명사만 추출하기, 결과값은 명사 리스트
    noun_list.append(nouns)
```

```python
text_df['nouns'] = noun_list
```

### 단순 토큰화 후 벡터 거리 구해서 유사 문장 찾기

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy as sp
```

```python
vectorizer = TfidfVectorizer(min_df=3, decode_error='ignore')
```

```python
#토큰화
t = Okt()
contents_tokens=[t.morphs(row) for row in text]
contents_for_vectorize = []
for content in contents_tokens:
  sentence=''
  for word in content:
    sentence = sentence + ' ' + word

    contents_for_vectorize.append(sentence)
#tf-idf 벡터화
X = vectorizer.fit_transform(contents_for_vectorize)
num_samples, num_features = X.shape
num_samples, num_features
```

```python
new_text = ['생체모방의 아름다움과 자연에서 영감을 받은 디자인 컨셉 축하']
```

```python
new_text_tokens = [t.morphs(row) for row in new_text]
new_text_for_vec = []

for content in new_text_tokens:
  sentence = ''
  for word in content:
    sentence = sentence + ' '+ word
  new_text_for_vec.append(sentence)
new_text_for_vec
```

```python
new_text_vec = vectorizer.transform(new_text_for_vec)
```

```python
from numpy import dot
from numpy.linalg import norm
import math
import numpy as np
```

```python
def dist_raw(v1,v2):
  delta = v1-v2
  return sp.linalg.norm(delta.toarray())

best_dist = -1 #그냥 설정함
best_i = None

for i in range(len(text)):
  #벡터화 한 프레임에서 1개씩 데려오기
  text_vec = X.getrow(i)
  d = dist_raw(text_vec, new_text_vec)

  print("--Text %i with dist=%.2f : %s" %(i,d,text[i]))

  if d > best_dist :
    best_dist = d
    best_i = i

print("--Best %i with dist=%.2f : %s" %(best_i, best_dist, text[best_i]))
```

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
```

```python
embedder = SentenceTransformer("jhgan/ko-sroberta-multitask")
```

```python
corpus_embeddings = embedder.encode(text, convert_to_tensor=True)
```

```python
# Query sentences:
queries = ['파도 이미지', '혁신적이고 지속 가능','육아 및 가족','하얀색 배경']
```

```python
text = list(text)
```

```python
# Find the closest 5 sentences of the corpus for each query sentence based on cosine similarity
for query in queries:
 query_embedding = embedder.encode(query, convert_to_tensor=True)
 cos_scores = util.pytorch_cos_sim(query_embedding, corpus_embeddings)[0]
 cos_scores = cos_scores.cpu()
 cos_df = pd.DataFrame(zip(text,cos_scores))
 cos_df.columns = ['text','cos_scores']
 cos_df = cos_df[cos_df['cos_scores']>=0.4]
 cos_df = cos_df.sort_values(by='cos_scores', ascending=False)
 
 print("\n\n======================\n\n")
 print("Query:", query)
 print("\nTop 5 most similar sentences in corpus:")
 print(cos_df[:20])
```
